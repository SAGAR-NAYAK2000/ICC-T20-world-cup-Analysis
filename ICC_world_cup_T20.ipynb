##Change object data coulums that are actually integer :

```python
categorical_dataFrame = crick_dataset.drop(['First Innings Score','Fall of wickets First Innings','Second Innings Score',
                                       'Second Innings Score','Fall of wickets Second Innings','Winning Margin',
                                       'Highest Score','Best Bowler Figure(Wickets Taken)','Best Bowler Figure(Runs Recieved)'],axis=1)
```

```python
categorical_dataFrame.head()
```

```python
worng_dtype  = crick_dataset[['First Innings Score','Fall of wickets First Innings','Second Innings Score',
                              'Fall of wickets Second Innings','Winning Margin','Highest Score',
                              'Best Bowler Figure(Wickets Taken)','Best Bowler Figure(Runs Recieved)']]
```

```python
numeric_data_frame = worng_dtype.apply(pd.to_numeric, errors='coerce')

numeric_data_frame.dtypes
```

```python
numeric_data_frame.describe()
```

```python
numeric_data_frame.head()
```

```python
# concat  categorical and casting numerical part
crick_dataFrame = pd.concat([categorical_dataFrame,numeric_data_frame],axis=1)
crick_dataFrame.head()
```

#venue according to matches:

```python
venue = crick_dataFrame.groupby(['Venue']).size().reset_index(name='matches')
venue
```

```python
ply.histogram(data_frame=venue,x ='Venue',y='matches',color_discrete_sequence=['#36256E'],text_auto=True)
```

#Interpretation:
##We can see the the vanue that played most is Kensington Oval, Bridgetown, Barbados.The venue is used (9) time in T20 world cup24.

##Second played venue is jointly (Nassau County International Cricket Stadium)and (Sir Vivian Richards Stadium, North Sound, Antigu).Both are played (8) time in their venue.

##Third played venue is jointly (Providence Stadium, Guyana) and (Daren Sammy National Cricket Stadium).Both are played (6) time in their venue.

#Match win based on toss win:

```python
toss_decision_vs_win = crick_dataFrame.groupby(['Toss Decision', 'Winners'])['Toss Decision'].size().reset_index(name='Win')

toss_decision_winn_count =  toss_decision_vs_win.groupby(['Toss Decision'])['Win'].sum().reset_index(name='Win')
toss_decision_winn_count
```

```python
ply.pie(data_frame=toss_decision_winn_count,
    names='Toss Decision',
    values='Win',
    color_discrete_sequence=['#133A75','#DA2C5A','#1C5D9C'],title='Toss wining desicion effect on win')
```

##Interpretation : We can see that after 76.4% of match win whom first choice was fielding after tossing. And 18.2% of match win whom first choice was batting after tossing. And 5.45% of total match was abandoned and some of them win by DLS method.

#Distribution of First Innings Score:

```python
score=crick_dataFrame[['First Innings Score','Second Innings Score']]
inining_score = score.dropna().reset_index(drop=True)
inining_score
```

```python
fig = ply.histogram(inining_score,
                   x='First Innings Score',
                   marginal='box',
                   nbins=10,
                   title='Distribution of First Innings Score',color_discrete_sequence=['#663399'])
fig.update_layout(bargap=0.4)
fig.show()
```

###Interpretation: We can see that,The 12 matches first innings score were in (100-119) range. The 9 matches first innings score were in (140-159) range. The 9 matches first innings score were in (180-199) range.

#Distribution of Second Innings Score:

```python
fig = ply.histogram(inining_score,
                   x='Second Innings Score',
                   marginal='box',
                   nbins=10,
                   title='Distribution of Second Innings Score',color_discrete_sequence=['#800080'])
fig.update_layout(bargap=0.4)
fig.show()
```

###Interpretation: We can see that,The 17 matches second innings score were in (100-119) range. The 9 matches second innings score were in (120-139) range. The 5 matches second innings score were in (60-79) range.

#Best bowling figure in a single match:

```python
bowler = crick_dataFrame[['Best Bowler','Best Bowler Figure(Wickets Taken)','Best Bowler Figure(Runs Recieved)']]
```

```python
bowler
```

note: In this dataset we find error in 'Best Bowler Figure(Wickets Taken)' columns .Nosthush Kenjige got 30 wicket and Alpesh Ramjani got 17 wicket in a single match. These are not a valid data. So, we should avoid them.

```python
best_bowling = bowler.sort_values(by='Best Bowler Figure(Wickets Taken)',ascending = False)
best_figure10  = best_bowling.iloc[2:12].reset_index(drop=True)
best_figure10
```

```python
df_melted = best_figure10.melt(id_vars='Best Bowler', value_vars=['Best Bowler Figure(Wickets Taken)', 'Best Bowler Figure(Runs Recieved)'], var_name='Variable', value_name='Value')


fig = ply.bar(df_melted, x='Best Bowler', y='Value', facet_col='Variable', color_discrete_sequence=['#C71585'],title='Best Bowling figure')
fig.show()

#When grouping with a length-1 list-like, you will need to pass a length-1 tuple to get_group in a future version of pandas. Pass `(name,)` instead of `name` to silence this warning.
```

#Player of the match got from the tournament :

```python
player_of_the_match = crick_dataFrame.groupby(['Player Of The Match']).size().reset_index(name ='count')
player_of_the_match = player_of_the_match.nlargest(6,columns='count').reset_index(drop=True)
player_of_the_match
```

```python
top_5_player = player_of_the_match.iloc[1:].reset_index(drop=True)
top_5_player
```

```python
sns.barplot(data=top_5_player,y ='Player Of The Match',x ='count' )
plt.title("Top5 player with 'Player Of The Match' award")
```
